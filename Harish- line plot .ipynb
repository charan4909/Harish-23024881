```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#line graph
df=pd.read_csv('covid.csv')
plt.plot(df['MEDICAL_UNIT'], df['AGE'], label='covid data')
plt.xlabel('SEX')
plt.ylabel('DATE_DIED')
plt.legend(loc="upper left")
plt.title('line graph for covid data', color='blue')
plt.show()

#print(df.isnull().sum())

df.head()
```
